```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics 
import numpy as np
```

##### We considered various parameters to determine driest month. We analysized data from various months with various parameters. 

##### Some of them being soil moisture, relative humidity, soil temperature etc.

```python
bay_area = pd.read_csv('question3_bay_area_weather.csv')
bay_area_equivalent = pd.read_csv('question3_bay_area_equivalent_weather.csv')
```

```python
bay_area['month'] = bay_area['month'].astype(str).str[:-2].astype(int)
bay_area_equivalent['month'] = bay_area_equivalent['month'].astype(str).str[:-2].astype(int)
```

```python
bay_area.head()
bay_area_equivalent.head()
```

```python
bay_area_average = bay_area.groupby(['month'], as_index=False).mean()
bay_area_equivalent_average = bay_area_equivalent.groupby(['month'], as_index=False).mean()
```

```python
bay_area_average
```

```python
fig, ax1 = plt.subplots(figsize=(10,8))

width = 0.35   
rng = np.random.RandomState(0)

ax1.bar(bay_area_average['month'], bay_area_equivalent_average['humidityAverage'], width,label='Bay Area Humidity',color = 'red',alpha = 0.6)
ax1.bar(bay_area_average['month']+ width , bay_area_equivalent_average['humidityAverage'], width,label='Equivalent Humidity',alpha = 0.7)
colors = rng.rand(6)
#sizes = 1000 * rng.rand(100)
area = (30 * np.random.rand(10))**2


ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('Average')  # we already handled the x-label with ax1
ax2.plot(bay_area_average['month'], bay_area_average['windAverage'], label='Bay Area Wind',linewidth = 3,marker ='o',color = 'yellow')
ax2.plot(bay_area_average['month'], bay_area_equivalent_average['windAverage'], label='Equivalent Wind',marker ='o',color = 'lightgreen')
ax2.plot(bay_area_average['month'], bay_area_average['airtempAverage'], label='Bay Area Air Temperature',linewidth = 3,marker ='o',color = 'cyan')
ax2.plot(bay_area_average['month'], bay_area_equivalent_average['airtempAverage'], label='Equivalent Air Temperature',marker ='o',color = 'darkgreen')
ax2.scatter(bay_area_average['month'], bay_area_average['precipitationAverage'],label= 'Bay Area Precipitation', c ='black',  marker = 'o')
ax2.scatter(bay_area_average['month'], bay_area_equivalent_average['precipitationAverage'],label= 'Equivalent Precipitation', c ='white',  marker = 'o')
ax2.legend(loc='upper left')
ax1.set_ylabel('Humidity')
ax1.set_xlabel('Years')
ax1.legend(loc='upper right')
plt.xticks(bay_area_average['month'] + width / 2, (2013, 2014, 2015, 2016, 2017))

plt.title('Comparison')
plt.savefig('question3_graph.png')
```

```python
fig, [[ax1, ax2],[ax3, ax4]] = plt.subplots(2, 2, figsize=(15,10))

width = 0.35   
rng = np.random.RandomState(0)

ax1.bar(bay_area_average['month'], bay_area_average['humidityAverage'], label="Humidity", color='red',align='center', alpha=0.6)
ax1.legend(loc="upper left")
ax1.set_ylabel('Humidity')
ax1.set_xlabel('Years')

ax2.plot(bay_area_average['month'], bay_area_average['windAverage'], label='Wind',linewidth = 3,marker ='o',color = 'darkgoldenrod')
ax2.legend(loc='upper left')
ax2.set_ylabel('Wind')
ax2.set_xlabel('Years')


ax3.plot(bay_area_average['month'], bay_area_average['airtempAverage'], label='Air Temperature',linewidth = 3,marker ='o',color = 'teal')
ax3.legend(loc='upper left')
ax3.set_ylabel('Air Temperature')
ax3.set_xlabel('Years')

ax4.plot(bay_area_average['month'], bay_area_average['precipitationAverage'], label='Precipitation',linewidth = 3,marker ='o',)
ax4.legend(loc='upper left')
ax4.set_ylabel('Precipitation')
ax4.set_xlabel('Years')

plt.title('Bay Area Weather')
plt.savefig('question3_bay_area_graph.png')
```
