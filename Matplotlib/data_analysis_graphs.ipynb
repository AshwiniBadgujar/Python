```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lowest_air_temp = pd.read_csv('lowest_air_temperature.csv')
lowest_surf_temp = pd.read_csv('lowest_surface_temperature.csv')
highest_surf_temp = pd.read_csv('highest_surface_temperature.csv')
highest_air_temp = pd.read_csv('highest_air_temperature.csv')
```

```python
lowest_air_temp.head()
lowest_surf_temp.head()
highest_surf_temp.head()
highest_air_temp.head()
```

```python
lowest_air_temp['month'] = lowest_air_temp['month'].astype(str).str[:-2].astype(int)
lowest_surf_temp['month'] = lowest_surf_temp['month'].astype(str).str[:-2].astype(int)
highest_surf_temp['month'] = highest_surf_temp['month'].astype(str).str[:-2].astype(int)
highest_air_temp['month'] = highest_air_temp['month'].astype(str).str[:-2].astype(int)
```

```python
lowest_air = lowest_air_temp.groupby(['month'])['temperature'].transform(min) == lowest_air_temp['temperature']
lowest_surf = lowest_surf_temp.groupby(['month'])['temperature'].transform(min) == lowest_surf_temp['temperature']
highest_surf = highest_surf_temp.groupby(['month'])['temperature'].transform(max) == highest_surf_temp['temperature']
highest_air = highest_air_temp.groupby(['month'])['temperature'].transform(max) == highest_air_temp['temperature']
```

```python
fig, axs= plt.subplots(nrows=2, ncols=1,figsize=(15,10))
axs[0].title.set_text('lowest Air Temperature')
axs[0].plot(lowest_air_temp[lowest_air]['month'], lowest_air_temp[lowest_air]['temperature'],linewidth=4)
axs[0].set_xticks([2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018])
axs[0].axvspan( 2010.7,2011.4,facecolor='b', alpha=0.2)
axs[0].set(xlabel='Years', ylabel='Air Temperature')
axs[1].title.set_text('Highest Air Temperature')
axs[1].plot(highest_air_temp[highest_air]['month'], highest_air_temp[highest_air]['temperature'],linewidth=4)
axs[1].set_xticks([2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018])
axs[1].axvspan( 2010.7,2011.4,facecolor='b', alpha=0.2)
axs[1].set(xlabel='Years', ylabel='Air Temperature')
plt.savefig('question1_graph1.png')
```

##### Now that we have minimum air temperature, we check the month from the graph. It looks like the month is approximately around 201100... To verify our answer, we go to the data and check the month with our lowest value.

```python
fig, axs= plt.subplots(nrows=2, ncols=1,figsize=(15,10))
axs[0].title.set_text('Lowest Surface Temperature')
axs[0].plot(lowest_surf_temp[lowest_surf]['month'], lowest_surf_temp[lowest_surf]['temperature'],linewidth=4,color = 'brown')
axs[0].set_xticks([2012,2013,2014,2015,2016,2017,2018])
axs[0].axvspan(2017.8,2018.2,facecolor='r', alpha=0.2)
axs[0].set(xlabel='Years', ylabel='Surface Temperature')
axs[1].title.set_text('Highest Surface Temperature')
axs[1].plot(highest_surf_temp[highest_surf]['month'], highest_surf_temp[highest_surf]['temperature'],linewidth=4,color = 'brown')
axs[1].set_xticks([2012,2013,2014,2015,2016,2017,2018])
axs[1].axvspan( 2014.8,2015.2,facecolor='r', alpha=0.2)
axs[1].set(xlabel='Years', ylabel='Surface Temperature')
plt.savefig('question1_graph2.png')
```
