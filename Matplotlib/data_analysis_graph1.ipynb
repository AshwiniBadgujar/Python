```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dryness = pd.read_csv('question2.csv')
```

```python
dryness.head()
```

```python
dryness['mn'] = dryness['month'].astype(str).str[4:].astype(int)
dryness['month'] = dryness['month'].astype(str).str[:-2].astype(int)
dryness_average = dryness.groupby(['month'], as_index=False).mean()
dryness
```

```python
fig, ax1 = plt.subplots(figsize=(10,8))

ax1.bar(dryness_average['month'],dryness_average['wetness'], label="Wetness", color='skyblue')
ax1.legend()
ax1.set_xlabel('Years')
ax1.set_ylabel('Wetness')
ax1.set_xticks([2011,2012,2013,2014,2015,2016,2017,2018,2019])
plt.title('Wetness')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('humidity')  # we already handled the x-label with ax1
ax2.plot(dryness_average['month'],dryness_average['humidity'],label="Humidity",linewidth=3,color = 'brown',marker='o')
ax2.legend(loc="upper left")

plt.savefig('question2_graph1.png')
```

```python

df = []
for region, sub_df in dryness.groupby('month',as_index=False):
    df.append(sub_df)
```

```python

months =['Jan','Feb','Mar','April','May','June','July','Aug','Sep','Oct','Nov','Dec']
```
